```python
import pandas as pd
import seaborn as sns
%matplotlib inline
```

https://archive.ics.uci.edu/ml/datasets/Iris

```python
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", 
                   header=None, # sem linha com header
                   names=["sepal length", "sepal width", "petal length", "petal width", "species"] # nome das colunas
                  )
```

```python
iris.describe().T[['min', 'max', 'mean', 'std']]
```

```python
sns.pairplot(iris, hue="species")
```

```python
iris.sample()
```

```python
species_ids, species_factor = pd.factorize(iris.species)

species_factor
```

```python
iris['species_id'] = species_ids

iris.head()
```
