# Line

```python
ts = pd.DataFrame(np.random.randn(1000, 4), 
               index=pd.date_range("1/1/2015", periods=1000),
               columns=list("ABCD"))
ts = ts.cumsum()
ts.plot.line(figsize=(10, 6))
```

# Bar/Barh

```python
tts = ts.sum()

tts
```

```python
tts.plot.bar()
# linha horizontal, na posi√ß√£o 0, cor blacK
plt.axhline(0, color='k')
```

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=list('abcd'))

df
```

```python
df.plot.bar(figsize=(10, 6))
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.div(df.sum(axis=1), axis=0).plot.barh(stacked=True)
plt.legend(bbox_to_anchor=(1., .5))
```

# Histogram

```python
df = pd.DataFrame({'a': np.random.randn(1000) + 1,
                   'b': np.random.randn(1000),
                   'c': np.random.randn(1000) - 1})

df.plot.hist(alpha=.5)
```

```python
df.plot.hist(stacked=True, bins=20)
```

# Scatter Matrix Plot

```python
from sklearn import datasets
from pandas.plotting import scatter_matrix
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['classe'] = iris.target
colors = df.classe.apply(lambda c: {0: 'b', 1: 'g', 2: 'r'}[c])

scatter_matrix(df.iloc[:, :-1], figsize=(10, 10), c=colors);
```
